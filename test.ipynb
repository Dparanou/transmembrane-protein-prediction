```python
import numpy as np
from Bio.PDB import *
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
```

```python
hydrophobic_amino_acid = ["PRO", "ALA", "VAL", "ILE", "LEU", "MET", "PHE", "TRP"]
```

```python
seq = []
atoms = []

for line in open('1qjp.pdb'):
  list = line.split()
  id = list[0]

  if id == 'SEQRES':
    seq.append(list[4:])
  if id == 'ATOM':
    atoms.append(np.float128(list[6:9]))
  
```

```python
X,Y,Z=np.array(atoms).T
ax = plt.axes(projection='3d')

ax.plot3D(X,Y,Z)
ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
plt.show()
```

```python
polarity = ""
for row in seq:
  for amino in row:
    if amino in hydrophobic_amino_acid:
      polarity += " Non"
    else:
      polarity += " Polar"
```
